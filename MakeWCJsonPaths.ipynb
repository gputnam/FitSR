```python
# Build the dict
output = {}

output["FieldResponse"] = {}
output["FieldResponse"]["axis"] = [1.0, 0.0, 0.0]
output["FieldResponse"]["origin"] = 0.
output["FieldResponse"]["tstart"] = 0.
output['FieldResponse']['period'] = 100.
output['FieldResponse']["speed"] = 0.0016

output['FieldResponse']['planes'] = [{} for _ in range(3)]

for p in range(3):
    output['FieldResponse']['planes'][p]['PlaneResponse'] = {}
    output['FieldResponse']['planes'][p]['PlaneResponse']['planeid'] = p
    output['FieldResponse']['planes'][p]['PlaneResponse']['location'] = 6 - 3*p
    output['FieldResponse']['planes'][p]['PlaneResponse']['pitch'] = 3.
    output['FieldResponse']['planes'][p]['PlaneResponse']["paths"] = []
    for ip, pitch in enumerate(pitches):
        output['FieldResponse']['planes'][p]['PlaneResponse']["paths"].append({})
        output['FieldResponse']['planes'][p]['PlaneResponse']["paths"][ip]["PathResponse"] = {}
        pathdir = output['FieldResponse']['planes'][p]['PlaneResponse']["paths"][ip]["PathResponse"]
        fname = filepath(p, pitch)
        
        current = []
        with open(fname) as f:
            for line in f:
                current.append(float(line.rstrip("\n")))
            
        pathdir["wirepos"] = 0
        pathdir["pitchpos"] = pitch
        pathdir["current"] = {}
        pathdir["current"]["array"] = {}
        pathdir["current"]["array"]["shape"] = [len(current)]
        pathdir["current"]["array"]["elements"] = current
```

```python
with open(OUTPUT, 'w') as f:
    json.dump(output, f)
```
